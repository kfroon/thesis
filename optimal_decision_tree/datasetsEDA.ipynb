```python
def load_data(df, split = True):
    X = df.iloc[:,:-1]
    y = df.iloc[:,-1:].squeeze()
    feature_names = X.columns
    if split == False:
        return X, y, feature_names
    if split == True:
        x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=390, test_size=0.3)
        return x_train, x_test, y_train, y_test, feature_names
```

```python
X,y, feature_names = load_data(df=breast_df, split=False)
```

```python
#Check corralation between features using pearson corrolation co-efficient
correlation = X.corr(method='pearson')
```

```python
fig, ax =plt.subplots(figsize=(6, 4))
plt.title("Correlation Plot")
sns.heatmap(correlation, mask=np.zeros_like(correlation, dtype=bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
plt.show()
```
